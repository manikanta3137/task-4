```python
import pandas as pd

df = pd.read_csv("student_placement_career_success_dataset.csv")

print(df.shape)
print(df.head())
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='placement_status', data=df)
plt.title("Placement Distribution")
plt.show()
```

```python
sns.boxplot(
    x='placement_status',
    y='cgpa',
    data=df
)

plt.show()
```

```python
sns.boxplot(
    x='placement_status',
    y='internships_completed',
    data=df
)

plt.show()
```

```python
sns.boxplot(
    x='placement_status',
    y='resume_score',
    data=df
)

plt.show()
```

```python
import numpy as np

numeric_df = df.select_dtypes(include=np.number)

corr = numeric_df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm')
plt.show()
```

```python
sns.histplot(df['aptitude_score'], kde=True)
plt.title("Aptitude Score Distribution")
plt.show()
```

```python
sns.histplot(df['communication_skills'], kde=True)
plt.title("Communication Skills Distribution")
plt.show()
```
